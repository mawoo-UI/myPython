1. 인덱싱 (Indexing) 연습
주어진 문자열 s = "Python Programming"에서 'g' 문자가 처음 나타나는 위치의 문자를 인덱싱을 사용하여 추출하고, 그 결과를 출력하세요.

2. 슬라이싱 (Slicing) 연습
주어진 문자열 data = "20250930_파이썬_기초"에서 날짜 부분 ("20250930")만 슬라이싱하여 추출하고 출력하세요.

```python
data = "20250930_파이썬_기초"

print(data[0:8])
```

3. 스텝 슬라이싱 (Step Slicing) 연습
주어진 문자열 sequence = "0123456789"를 슬라이싱을 사용하여 역순으로 뒤집어서 출력하세요.

```python
sequence = "0123456789"
print(sequence[::-1])
```

4. 문자열 메서드 - split() 및 join() 연습
주어진 문자열 sentence = "Hello world, this is a test."를 공백을 기준으로 **분리(split)**한 후, 분리된 단어들을 다시 **쉼표와 공백 (", ")**으로 연결하여 하나의 문자열로 만들고 출력하세요.

```python
sentence = "Hello world, this is a test."
print(sentence.replace(", ", " ").replace(" ", ", "))
```

5. 문자열 메서드 - replace() 및 upper() 연습
주어진 문자열 text = "learning python is fun and rewarding"에서 다음 두 작업을 순서대로 수행하고 최종 결과를 출력하세요.

"python" 이라는 단어를 **"JAVA"**로 **교체(replace)**하세요.

결과 문자열 전체를 **대문자(upper)**로 변환하세요.

```python
text = "learning python is fun and rewarding"
print(text.replace("python", "java").upper())
```
