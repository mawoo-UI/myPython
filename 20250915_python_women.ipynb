## [자동 판매기 프로그램]


```python
# 자동 판매기 프로그램

자동판매기는 사용자로부터 투입한 돈과 물건 값을 입력받는다.
물건값은 100원 단위라고 가정한다. 프로그램은 잔돈을 계산하여 출력한다.
자판기는 동전 500원,100원 짜리만 가지고 있다고 가정한다.
투입한 금액과 물건값을 입력받아 거스름돈으로 500원 동전 몇개와 100원 동전 몇개를 내어 보내면 되는지 계산하는 프로그램을 작성하라

money = int(input("투입한 돈: "))
price = int(input('물건값: '))

change = money - price
print("거스름돈: ", change)
coin500s = change // 500
change = change % 500
coin100s = change // 100

print("500원 동전의 개수", coin500s)
print("100원 동전의 개수", coin100s)
```

```python
# 숫자 맞히기 게임 (UP/DOWN 게임) - 기회는 10번! (1)
import random
com = random.randint(1,20)

for i in range(1,11):
    print("게임", i, "회", end = " : ")
    user = int(input("숫자를 맞춰봐 (1-20): "))
    
    if com == user:         # 컴퓨터와 내가 생각하는 수가 같다면
        print("정답입니다. 축하합니다!!")
        break
    else:                   # 컴퓨터와 내가 생각하는 수가 같지 않다면
        if com > user:          # 컴퓨터가 생각하는 수가 내가 생각하는 수 보다 크다면
            print("{} , UP! 다시 맞춰봐...".format(user))
            continue
        else:                   # 컴퓨터가 생각하는 수가 내가 생각하는 수 보다 작다면
            print("{} , DOWN! 다시 맞춰봐...".format(user))
            continue
        # print("정답이 아닙니다")
        
print("게임 종료")
```

## [D-day 계산 프로그램]

```python
# D-day 계산 프로그램
from datetime import datetime, timedelta

def today():
    '''현재 날짜를 구해서 반환 해주는 함수
    반환값 : 현재 datetime '''
    cur_date = datetime.now()
    return cur_date

def dday(target_date):
    '''오늘로부터 D-day를 구해서 남은 날짜를 반환 해주는 함수
    입력값 : D-day를 구하고자 하는 날짜
    반환값 : None '''
    cur_date = datetime.now()
    delta = target_date - cur_date
    # print(delta)
    days_left = delta.days
    print(f"{target_date}, D-day :{days_left}일 남았습니다")

def change_date():
    '''키보드로 입력받은 문자열을 날짜로 변경해주는 함수
    입력값 : None
    반환값 : 키보드로 입력받은 문자열의 (년, 월, 일) '''
    t_day = input("타겟 날짜를 입력하세요(년 월 일) :")
    t_dayList = t_day.split()
    t_dayList[0] = int(t_dayList[0])
    t_dayList[1] = int(t_dayList[1])
    t_dayList[2] = int(t_dayList[2])
    return datetime(t_dayList[0],t_dayList[1],t_dayList[2])

dday(change_date())
```

```python
# 직접 전처리 (파싱)
# s = '{"name": "Joy", "age": 45, "is_student": false}'.replace("false", "False")

def simple_json_parser(s):
    s = s.strip().strip("{}")   # 중괄호 제거
    items = s.split(",")
    result = {}
    for item in items:
        key, value = item.split(":")
        key = key.strip().strip('"')
        value = value.strip().strip('"')
        result[key] = value
    return result

json_str = '{"name": "Joy", "age": "45"}'
data = simple_json_parser(json_str)
print(data)
print(type(data))
```

```python
# 클래스 상속
# 부모 클래스(super) - Animal
# 자식 클래스(sub) - Dog, Cat
class Animal:
    def __init__(self):
        print("동물 태어남 : ", end="")    
    def sound(self):
        print("동물 소리 : ", end="")
    def sleep(self):
        print("Zzzzz.....")
        
class Dog(Animal):
    def __init__(self):
        # super().__init__()    # 상속의 상속, 다중상속 시 한꺼번에 호출
        print("강아지")
    def sound(self):
        super().sound()
        print("멍멍")
        
class Cat(Animal):
    def __init__(self):
        Animal.__init__(self)   # 부모클래스 하나만 호출
        print("고양이")
    def sound(self):
        Animal.sound(self)
        # Dog.sound(self)       # 부모클래스가 아닌경우는 사용 불가
        print("야옹")
```

```python
# 객체 생성
d = Dog()
c = Cat()
```

```python
# 자식 클래스의 메서드와 상속받은 부모 클래스의 메서드 사용
d.sound()
d.sleep()

c.sound()
c.sleep()
```

```python
# 부모 클래스 : 스쿨멤버
# 자식 클래스 : 선생님, 학생
class SchoolMember:
    '''학교 멤버 정보를 관리하는 새로운 자료형(클래스) 입니다'''
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print("새 멤버:", end =" ")
        
    def tell(self):
        print(f"이름: {self.name} 나이: {self.age}", end=" ")
        
class Teacher(SchoolMember):
    '''선생님 정보 새로운 자료형(클래스) 입니다'''
    def __init__(self, name, age, salary):
        SchoolMember.__init__(self, name, age)
        self.salary = salary
        # print(f'self.salary ID:{id(self.salary)}, salary ID:{id(salary)}')
        print(f"New 선생님 - {self.name}")
    def tell(self):
        SchoolMember.tell(self)
        print("연봉:",self.salary)
        
class Student(SchoolMember):
    '''학생 정보 새로운 자료형(클래스) 입니다'''
    def __init__(self, name, age, score):
        SchoolMember.__init__(self, name, age)
        self.score = score
        print(f"New 학생 - {self.name}")
    def tell(self):
        SchoolMember.tell(self)
        print("학점:",self.score)
```

```python
t = Teacher('Joy', 45, 5000)
s = Student('Loy', 20, 3.8)
```

```python
t.tell()
s.tell()
```

```python
# 클래스 : 객체 캡슐화 (변수 + 메서드)
# 클래스가 갖는 필드(변수)의 속성 - public, protected, private
# 변수명 : public - 클래스 내, 같은 패키지, 자식클래스, 다른패키지 접근 가능
# _변수명 : protected - 클래스 내, 같은 패키지, 자식 클래스 접근 가능 (다른패키지 불가)
# __변수명 : private - 클래스 내에서만 접근 가능 (name mangling - 클래스 밖에서 접근 불가)

class Account:
    def __init__(self, name, balance):
        self.name = name            # public 멤버
        self.__balance = balance    # private 멤버 , 이름 맹글링(name mangling) 
        self._branch = '동대문'      # protected 멤버
    def deposit(self, money):
        self.__balance += money     # private 멤버 , 이름 맹글링(name mangling)
    def inquire(self):
        print(f"{self.name}의 잔액은 {self.__balance:,}원입니다.")
```

```python
customer1 = Account('Joy', 5000)
customer2 = Account('Jeff', 7000)
```

```python
customer1.deposit(5000)
```

```python
customer1.inquire()
customer2.inquire()
```

```python
# 파일 암호화 저장 프로그램
public_key = 100
secure_f = None
txt_str, sec_str = '', ''

secure_f = open("secure.txt", 'w')

while True:
    txt_str = input('전달할 메시지 :')
    if txt_str == '':
        break
    for ch in txt_str:
        sec_str += chr(ord(ch)+public_key)    # 암호화 키 값 더하기    
    sec_str += chr(ord('\n')+public_key)

# 문자열 저장
secure_f.writelines(sec_str)
secure_f.close()

print("암호 파일 저장 완료")
```

```python
# 암호파일 해독 프로그램
public_key = 100
secure_f, decode_f = None, None
txt_str, sec_str = '', ''

secure_f = open("secure.txt", 'r', encoding='UTF-8')
decode_f = open("decoded.txt", 'w', encoding='UTF-8')

# 파일에서 문자열로 읽기
sec_str = secure_f.readline()
secure_f.close()

for ch in sec_str:
    # print(ch)
    txt_str += chr(ord(ch)-public_key)   # 암호화 키 값 빼기
print(txt_str)

# 문자열 파일에 저장하기
decode_f.writelines(txt_str)
decode_f.close()
print("해독 파일 저장 완료")
```
