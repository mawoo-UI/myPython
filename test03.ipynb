문자열을 입력받아 다음을 수행하는 함수 analyzeString을 작성하세요:

```python
def analyzeString(string_value):
  if  string_value == "":
    print("빈 문자열")
  elif len(string_value) <= 5:
      print("짧은 문자열")
  elif 6 <= len(string_value) <= 15:
      print("보통 문자열")
  else:
      print("긴 문자열")
string_value = input("문자열을 입력하세요:> ")
analyzeString(string_value)
```

복합 함수
나이: 14세 이상 120세 이하
이메일: "@"가 포함되어야 함
비밀번호: 8자 이상이어야 함

```python
def checkSignup(user):

  if not (14 <= user['age'] <= 120):
      print("나이가 조건에 맞지 않습니다. (14~120세)")
  elif '@' not in user['email']:
      print("이메일에 @가 포함되어야 합니다.")
  elif len (user['password']) < 8 :
      print("비밀번호가 8자 이상이어야 합니다.")
  else:
      print("가입이 즉시 가능합니다.")

user_info = {
  'age': int(input("가입자 나이를 입력하세요.")),
  'email' : input("이메일을 입력해주세요(@ 가 포함되어야합니다.)"),
  'password' : input("비밀번호를 입력해주세요(8자 이상):> ") 
}
checkSignup(user_info)
```

고차함수 활용
숫자 리스트와 연산 타입을 받아서 결과를 반환하는 함수 calculate를 작성하세요.

```python
def calculate(numbers, op):

  if op == 'sum':
    print(" + ".join(str(n) for n in numbers), "=", end=' ')
    return sum(numbers)
  elif op == 'max':
    print("입력 값: ".join(str(n) for n in numbers), "=" ,end= ' ')
    print("최대값:", max(numbers))
    return max(numbers)
  elif op == 'min':
    print("입력 값: ".join(str(n) for n in numbers), "=" ,end= ' ')
    print("최소값:", min(numbers))
    return min(numbers) 
  else:
    return "지원하지 않는 연산입니다."

count = input("숫자들을 입력하세요(공백으로 구분):> ")
numbers = [int(x) for x in count.split()]
op = input("연산을 선택하세요 (sum, max, min):> ")
result = calculate(numbers, op)
print("결과:", result)
```

온라인 쇼핑몰의 할인 계산기를 만드세요. 함수 calculateDiscount는 다음을 고려해야 합니다:
매개변수:

상품 가격 (price)
고객 등급 (grade): "bronze", "silver", "gold", "platinum"
구매 수량 (quantity)


```python
def calculate(price, grade, quantity):

  if grade == "silver":
    discount = 0.05
  elif grade == "gold":
    discount = 0.10
  elif grade == "platinum":
    discount = 0.15
  else:
    discount = 0.0

  total = price * quantity
  total = total - (1 - discount)



  if total >= 10:
    total = total - 0.95
  
  if total >= 10000:
    total = total - 0.97
  
  return int(total)

price = int(input("상품 가격을 입력하세요:> "))
print("고객 등급을 입력하세요 (bronze, silver, gold, platinum): ")
grade = input("등급 입력: ")
quantity = int(input("구매 수량을 입력하세요 "))

# 이후 grade를 함수에 전달
final_price = calculate(price, grade, quantity)
print("최종 결제 금액: {:,} 원".format(final_price))
```
