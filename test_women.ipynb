3.4 조건문 문제 (학점 계산 프로그램)

```python
# 1. 점수를 키보드로부터 입력값을 받습니다 (0 점 ~ 100 점)

# 2. 점수값에 따라 학점을 출력해 줍니다
#    90 점 이상 100 점 이하 : A
#    80 점 이상 : B
#    70 점 이상 : C
#    60 점 이상 : D
#    60 점 미만 : F
#    그외의 값 : 잘못된 값입니다.
```

```python
while True:
    x = int(input("점수를 입력하세요"))

    if 90 <= x <= 100:
        grade = 'A'
        print(f"당신의 점수는 {x}입니다")
        print("당신의 학점은 "+grade+"입니다")
        break
    elif 80 <= x < 90 :
        grade = 'B'
        print("당신의 점수는 {}입니다".format(x))
        print("당신의 학점은 "+grade+"입니다")
        break
    elif 70 <= x < 80 :
        grade = 'C'
        print("당신의 점수는 {}입니다".format(x))
        print("당신의 학점은 "+grade+"입니다")
        break
    elif 60 <= x < 70  :
        grade = 'D'
        print("당신의 점수는 {}입니다".format(x))
        print("당신의 학점은 "+grade+"입니다")
        break
    elif 0 <= x < 60  :
        grade = 'F'
        print("당신의 점수는 {}입니다".format(x))
        print("당신의 학점은 "+grade+"입니다")
        break
    else:
        print("잘못된 점수 값입니다. 다시 입력하세요")
```

10.3 데코레이터(@)

```python
class Robot:
    population = 0
    
    def __init__(self, name):
        self.name = name
        print("Initializing...{}".format(self.name))
        Robot.population += 1
    
    def die(self):
        print("{} is destroying...".format(self.name))
        Robot.population -= 1
        
        if Robot.population == 0:
            print("{} 로봇이 마지막 로봇 이었습니다. 현재 {:d}대 남아있습니다".format(self.name, Robot.population))
        else:
            print("{} 로봇이 파기되었습니다. 현재 {:d}대 남아있습니다".format(self.name, Robot.population))
    
    def say_hi(self):
        print("Hello, My name is {}".format(self.name))
    
    @classmethod # 데코레이터
    def how_many(cls):
        print("현재 로봇이 {:d}대 입니다.".format(cls.population))
    
droid1 = Robot("R001")
droid2 = Robot("R002")
droid1.say_hi()  # 객체.메서드  
droid2.say_hi()
Robot.how_many() # 클래스.메서드
droid1.die()
Robot.how_many() # 클래스.메서드
droid2.die()



```

10.4 클래스 상속 (inheritance)

```python
class SchoolMember:
    def __init__(self,name,age):
        self.name = name
        self.age = age
        print(f"스쿨멤버가 생성되었습니다 : {self.name}")
    def tell(self):
        print(f"이름 : {self.name}, 나이 : {self.age}", end=" ")

class Teacher(SchoolMember):
    def __init__(self,name,age,salary):
        SchoolMember.__init__(self,name,age)
        self.salary = salary
        print(f"선생님 입니다 : {self.name}")
    def tell(self):
        SchoolMember.tell(self)
        print(f"연봉 : {self.salary}")

class Student(SchoolMember):
    def __init__(self, name, age, score):
        SchoolMember.__init__(self, name, age)
        self.score = score
        print(f"학생 입니다 : {self.name}")
    def tell(self):
        SchoolMember.tell(self)
        print(f"학점 : {self.score}")
        
        
t = Teacher("Mr. Kim", 50, 100000000)
s = Student("Loy", 25, 4.3)

t.tell()
s.tell()
```

11.2 피클링(pickling)

```python
import pickle
a = [100,200,True,"Hi",[10,5.5]]

f = open("myListPickle", 'w+b')
pickle.dump(a, f)
f.close()
```

```python
import pickle

f = open("myListPickle", 'r+b')
b = pickle.load(f)
f.close()

print(b)
```
