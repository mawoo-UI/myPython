```python
phone_num = {'Emily':'010-1234-5678','Eric':'010-2725-6789','Ellie':'010-3456-789' }

print(phone_num)

for name in phone_num: #key만 출력
  print(name)

print()

for x in phone_num.keys():  #key만 출력
  print(x)

print()


for x in phone_num.values(): #value만 출력
  print(x)


for x in phone_num.items(): # (key, value) 출력
  print(x)

print()

for x, y in phone_num.items(): #key, value 를 각각의 변수로 출력
  print(x, y, sep=":")


```

```python
a = 1, 2, 3, 4 # 튜플로 인식함()

print(a)
print(a[0]) #시퀀스형 데이터라 가능함
# a[0] = 5 # 튜플은 값 변경 불가 /리스트와 다름

v1 , v2 = 2, 3

print(v1, v2)


v1, *v2 = 11, 22, 33, 44  #r 벨류가 더 많으면 원래 실행안되지만 * 사용시엔 리스트로 바꿔줌
print(v1, v2)
type(v1)
type(v2)

*v1, v2 = 11, 22, 33, 44 
print(v1, v2)


v1, *v2, v3 = 11, 22, 33, 44 
print(v1, v2, v3)


*v1, v2, v3 = 11, 22, 33, 44 
print(v1, v2, v3)
```

```python
def calc_temperature(kelvvin):
  cel = kelvvin - 273
  fah = cel * 9 /5 + 32
  return cel , fah

c, f = calc_temperature(300)
print(c,"C", f, "F")


print(type(calc_temperature)) #함수도 클래스로 인식
```

바다코끼리 연산자(the walrus operator) v3.8이상 지원

```python
# :=
name = "This example sentence"
if(n := len(name)) > 3:
  print(name + "의 길이는 " + str(n) + "이다.")
```

match(switch문과 비슷) v3.10이상 지원

```python
x = 2
match x:
  case 1:
    print("one")
  case 2:
    print("two")
```

```python
grade = input("학년을 입력하세요 : ")

match grade:
  case '1' | '2' | '3':
    print(grade + "학년은 저학년 입니다.")
  case '4' | '5' | '6':
    print(f"{grade} + 학년은 고학년 입니다.")
  case _ : #_ 는 와일드카드(앞에 조건외의 다른 경우)
    print("1 ~ 6 까지만 입력 가능!")
```

```python
grade = input("학년을 입력하세요 : ")

if  ('1' or '2' or '3'):
  print(grade + "학년은 저학년 입니다.")
elif ('4' or '5' or '6'):
  print(f"{grade} + 학년은 고학년 입니다.")
else:
  print("1 ~ 6 까지만 입력 가능!")
```
