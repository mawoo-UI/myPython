```python
class Person():
  # pass
  def __init__(self, name): #따로 호출하지않아도 자동으로 호출되는 메소드(생성자)
    self.name = name

  def say_hi(self):
    print("My name is", self.name)
    # print("Hello, how are you?", self.name)
  
  def say_bye(self):
    print("Good bye,", self.name)

p = Person(name="Christin")

p.say_hi()
p.say_bye()

# __ 00__ 언더바 2개는 특수한거에 쓰임
```

```python
class Robot:
  population = 0

  def __init__(self, name):
    self.name = name
    # print(f"initalizing...{self.name}")
    print("initalizing...{}".format(self.name))
    Robot.population += 1

  def die(self):
    print("{} is destroying.".format(self.name))  
    Robot.population -= 1

    if Robot.population == 0:
        print("{}로봇이 마지막 로봇 이었습니다. 현재 {}대 남아있습니다.".format(self.name, Robot.population))
    else:
        print("{}로봇이 파기되었습니다. 현재 {}대 입니다.".format(self.name , Robot.population))  


  def say_hi(self):
    print(f"Hello, My name is {self.name}")

  @classmethod #데코레이터
  def how_many(cls):
    print("현재 로봇이 {:d}대 입니다.".format(cls.population))



droid1 = Robot("R001")
droid2 = Robot("R002")
droid1.say_hi() # 객체.메서드로 접근
droid2.say_hi()
Robot.how_many() #클래스.메서드로 접근

droid1.die()
droid2.die()




```

상속

```python
class SchoolMember:
  def __init__(self, name, age):
    self.name = name
    self.age = age
    print("스쿨멤버가 생성되었습니다 : {}".format(self.name))
    # print(f"스쿨멤버가 생성되었습니다 : {self.name}")
  def tell(self):
    print(f"이름: {self.name}, 나이: {self.age}", end=" " )


class Teacher(SchoolMember):
  def __init__(self, name, age,salary):
    SchoolMember.__init__(self, name, age)
    self.salary = salary
    print(f"선생님 입니다 : {self.name}")

  def tell(self):
    SchoolMember.tell(self)
    print(f"연봉 : {self.salary}")

class Student(SchoolMember):
  def __init__(self, name, age, score):
    SchoolMember.__init__(self, name, age)
    self.score = score
    print(f"학생 입니다 : {self.name}")

  def tell(self):
    SchoolMember.tell(self)
    print(f"학점 : {self.score}")
  

t = Teacher("Mr. Kim", 50, 1000000000)
s = Student("Loy", 25, 4.3)

t.tell()  #상속 오버라이딩
s.tell()
```

try / exccept  : 예외 처리 (에러메세지 표시)

```python
try:
  f = open("test.txt", 'r')
  str = f.read()
  f.close()

  print(str)
  
except IOError:
  #에러 종류에 따라서 정의
  print("파일을 찾을 수 없습니다.")
```

에러와 예외의 계층 구조

```python
a = [1, 2, 3]

try:
  val = a[10]
  print("val=", val)
except:
  print("에러발생!")
else:
  print("에러없이 동작 완료!")
finally:
  print("항상 출력됨")
```

Exception 클래스와 as 키워드

```python
a = [1, 2, 3]

try:
  val = a[10]

except Exception as e:
  print("에러발생!")
  print(e)
```
